# ARACNe output adapter

```python
import os
import sys
import glob
import pandas as pd
import numpy as np
```

```python
example_out_dir = "/cellar/users/aklie/projects/igvf/topic_grn_links/grn_inference/aracne/results/Bridge_Satpathy/balanced_genotype_microglia/0.05/raw/B6J"
```

```python
adj_tsvs = glob.glob(os.path.join(example_out_dir, "**", "network.txt"))
```

```python
adj = pd.DataFrame()
for adj_tsv in adj_tsvs:
    adj_df = pd.read_csv(adj_tsv, sep="\t")
    adj_df["aracne_regulator_type"] = os.path.basename(os.path.dirname(adj_tsv))
    adj = pd.concat([adj, adj_df])
adj.head()
```

```python
adj.rename(
    columns={
        "Target": "target",
        "MI": "weight_unsigned",
        "pvalue": "p"},
    inplace=True
)
```

```python
adj["weight_signed"] = np.nan
```

```python
adj["weight_minmax_normalized"] = (adj["weight_unsigned"] - adj["weight_unsigned"].min()) / (adj["weight_unsigned"].max() - adj["weight_unsigned"].min())
```

```python
adj["description"] = np.nan
```

```python
adj["-logp"] = -np.log10(adj["p"])
```

```python
adj.reset_index(inplace=True)
adj.rename(columns={"Regulator": "source"}, inplace=True)
```

```python
adj = adj[["source", "target", "weight_signed", "weight_unsigned", "weight_minmax_normalized", "p", "-logp", "description", "aracne_regulator_type"]]
```
